```python
import requests
import json

url = "https://raw.githubusercontent.com/KajPPancho/Final-Project-CS2/refs/heads/main/data/ecommerce.json"
response = requests.get(url)

orders = json.loads(response.text)

print("JSON file loaded")

ref = db.reference("orders")

for order in orders:
    ref.child(str(order["order_id"])).set(order)

print("Data uploaded successfully!")
```

```python
# Feature 1

orders = db.reference("orders").get()


if isinstance(orders, dict):

    for order_id, order in orders.items():
        print(f"Order {order_id} Revenue:", order["total_amount"])
elif isinstance(orders, list):

    for order in orders:
        if order:
            print(f"Order {order['order_id']} Revenue:", order["total_amount"])
else:
    print("No orders found.")


```

```python
# Feature 2
orders = db.reference("orders").get()

if orders is None:
    print("No orders found in Firebase.")
else:
    for order in orders:
        if order:
            most_expensive = max(order["items"], key=lambda x: x["price"])
            print(f"Order {order['order_id']} - Most Expensive Item:", most_expensive)
```

```python
# Feature 3
orders = db.reference("orders").get()

if orders:
    for order in orders:
        if order:
            cheapest = min(order["items"], key=lambda x: x["price"])
            print(f"Order {order['order_id']} - Cheapest Item:", cheapest)
else:
    print("No orders found in Firebase.")
```

```python
# Feature 4
orders = db.reference("orders").get()

if orders:
    customers = {order["customer"] for order in orders if order}
    for customer in customers:
        total_spent = sum(o["total_amount"] for o in orders if o and o["customer"] == customer)
        print(f"Customer {customer} - Total Spent:", total_spent)
else:
    print("No orders found in Firebase.")
```

```python
# Feature 5
orders = db.reference("orders").get()

if orders:
    product_counts = {}
    for order in orders:
        if order:
            for item in order["items"]:
                product_counts[item["product"]] = product_counts.get(item["product"], 0) + item["quantity"]

    most_ordered = max(product_counts, key=product_counts.get)
    least_ordered = min(product_counts, key=product_counts.get)

    print("Most Ordered Product:", most_ordered)
    print("Least Ordered Product:", least_ordered)
else:
    print("No orders found in Firebase.")
```

```python
# Feature 6
orders = db.reference("orders").get()

if orders:
    customer_counts = {}
    for order in orders:
        if order:
            customer_counts[order["customer"]] = customer_counts.get(order["customer"], 0) + 1

    multi_orders = [cust for cust, count in customer_counts.items() if count > 1]

    print("Customers with Multiple Orders:", multi_orders)
    if multi_orders:
        print("Customers with Multiple Orders:", multi_orders)
    else:
        print("No customers with multiple orders.")
else:
    print("No orders found in Firebase.")
```

```python
# Planned Outputs

orders = db.reference("orders").get()

if orders:
    orders = db.reference("orders").get()

    if orders:
        product_counts = {}
        for order in orders:
            if order:  # skip None entries
                for item in order["items"]:
                    product_counts[item["product"]] = product_counts.get(item["product"], 0) + item["quantity"]

        if product_counts:
            most_sold = max(product_counts, key=product_counts.get)
            least_sold = min(product_counts, key=product_counts.get)

            print("Most Sold Product:", most_sold)
            print("Least Sold Product:", least_sold)
        else:
            print("No products found in orders.")
    else:
        print("No orders found in Firebase.")

    avg_products = sum(
        sum(item["quantity"] for item in order["items"])
        for order in orders if order
    ) / len(orders)


    customer_counts = {}
    for order in orders:
        if order:
            customer_counts[order["customer"]] = customer_counts.get(order["customer"], 0) + 1

    avg_spend = sum(order["total_amount"] for order in orders if order) / len(customer_counts)


    customers_in_timeframe = len({
        order["customer"] for order in orders if order and "2025-08-01" <= order["order_date"] <= "2025-08-05"
    })

    print("Average Products per Order:", avg_products)
    print("Average Spend per Customer:", avg_spend)

    if customers_in_timeframe > 0:
        print("Customers served between 2025-08-01 and 2025-08-05:", customers_in_timeframe)
    else:
        print("No customers served in that timeframe.")
else:
    print("No orders found in Firebase.")
```

```python
while True:
    print("\n===== ORDER DATABASE MENU =====")
    print("1. Display Orders")
    print("2. Add Order")
    print("3. Update Order Status")
    print("4. Delete Order")
    print("5. Features Menu")
    print("0. Exit")

    choice = input("Enter choice: ")

    # DISPLAY
    if choice == "1":
        print("\nOrder List:")
        for order in orders:
            print(f"\nOrder ID: {order['order_id']}")
            print(f"Customer: {order['customer']}")
            print(f"Date: {order['order_date']}")
            print(f"Status: {order['status']}")
            print(f"Total: {order['total_amount']}")

            print("Items:")
            for item in order["items"]:
                print(f" - {item['product']} ({item['quantity']}x)")

    # ADD
    elif choice == "2":
        try:
            order_id = int(input("Enter Order ID: "))
            customer = input("Enter Customer Name: ")
            date = input("Enter Order Date (YYYY-MM-DD): ")
            status = input("Enter Status: ")
            payment = input("Enter Payment Method: ")
            address = input("Enter Shipping Address: ")

            items = []
            while True:
                add_item = input("Add item? (y/n): ")
                if add_item.lower() != "y":
                    break

                product = input("Product: ")
                category = input("Category: ")
                price = float(input("Price: "))
                quantity = int(input("Quantity: "))

                items.append({
                    "product": product,
                    "category": category,
                    "price": price,
                    "quantity": quantity
                })

            total = sum(i["price"] * i["quantity"] for i in items)

            orders.append({
                "order_id": order_id,
                "customer": customer,
                "order_date": date,
                "status": status,
                "payment_method": payment,
                "shipping_address": address,
                "items": items,
                "total_amount": total
            })

            print("Order added successfully!")

        except:
            print("Invalid input!")

    # UPDATE
    elif choice == "3":
        oid = int(input("Enter Order ID to update: "))
        found = False
        for order in orders:
            if order["order_id"] == oid:
                new_status = input("Enter new status: ")
                order["status"] = new_status
                print("Order updated successfully!")
                found = True
                break
        if not found:
            print("Order not found.")

    # DELETE
    elif choice == "4":
        oid = int(input("Enter Order ID to delete: "))
        found = False
        for order in orders:
            if order["order_id"] == oid:
                orders.remove(order)
                print("Order deleted successfully!")
                found = True
                break
        if not found:
            print("Order not found.")

    # FEATURES SUB-MENU
    elif choice == "5":
        while True:
            print("\n===== FEATURES MENU =====")
            print("1. Calculate total revenue of an order")
            print("2. Find the most expensive item in the order")
            print("3. Find the cheapest item in the order")
            print("4. Show the total amount spent by customer")
            print("5. Calculate most ordered product by quantity")
            print("6. Track customers with multiple orders")
            print("0. Back to Main Menu")

            fchoice = input("Enter feature choice: ")

            if fchoice == "1":
                oid = int(input("Enter Order ID: "))
                for order in orders:
                    if order["order_id"] == oid:
                        print(f"Total Revenue: {order['total_amount']}")
                        break

            elif fchoice == "2":
                oid = int(input("Enter Order ID: "))
                for order in orders:
                    if order["order_id"] == oid:
                        expensive = max(order["items"], key=lambda x: x["price"])
                        print(f"Most Expensive Item: {expensive['product']} (${expensive['price']})")
                        break

            elif fchoice == "3":
                oid = int(input("Enter Order ID: "))
                for order in orders:
                    if order["order_id"] == oid:
                        cheap = min(order["items"], key=lambda x: x["price"])
                        print(f"Cheapest Item: {cheap['product']} (${cheap['price']})")
                        break

            elif fchoice == "4":
                customer = input("Enter Customer Name: ")
                total_spent = sum(order["total_amount"] for order in orders if order["customer"] == customer)
                print(f"Total Spent by {customer}: ${total_spent}")

            elif fchoice == "5":
                product_counts = {}
                for order in orders:
                    for item in order["items"]:
                        product_counts[item["product"]] = product_counts.get(item["product"], 0) + item["quantity"]
                if product_counts:
                    most_ordered = max(product_counts, key=product_counts.get)
                    print(f"Most Ordered Product: {most_ordered} ({product_counts[most_ordered]} units)")
                else:
                    print("No products found.")

            elif fchoice == "6":
                customer_orders = {}
                for order in orders:
                    customer_orders[order["customer"]] = customer_orders.get(order["customer"], 0) + 1
                multiple = [c for c, count in customer_orders.items() if count > 1]
                print("Customers with multiple orders:", multiple if multiple else "None")

            elif fchoice == "0":
                break

            else:
                print("Invalid feature choice.")

    # EXIT
    elif choice == "0":
        with open("ecommerce.json", "w") as f:
            json.dump(orders, f, indent=2)
        print("Exiting program...")
        break

    else:
        print("Invalid choice.")
```
