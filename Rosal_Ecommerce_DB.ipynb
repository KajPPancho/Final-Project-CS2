```python
import requests
import json

url = "https://raw.githubusercontent.com/KajPPancho/Final-Project-CS2/refs/heads/main/data/ecommerce.json"
response = requests.get(url)

orders = json.loads(response.text)

print("JSON file loaded")

ref = db.reference("orders")

for order in orders:
    ref.child(str(order["order_id"])).set(order)

print("Data uploaded successfully!")
```

```python
# Feature 1

orders = db.reference("orders").get()


if isinstance(orders, dict):

    for order_id, order in orders.items():
        print(f"Order {order_id} Revenue:", order["total_amount"])
elif isinstance(orders, list):

    for order in orders:
        if order:
            print(f"Order {order['order_id']} Revenue:", order["total_amount"])
else:
    print("No orders found.")


```

```python
# Feature 2
orders = db.reference("orders").get()

if orders is None:
    print("No orders found in Firebase.")
else:
    for order in orders:
        if order:
            most_expensive = max(order["items"], key=lambda x: x["price"])
            print(f"Order {order['order_id']} - Most Expensive Item:", most_expensive)
```

```python
# Feature 3
orders = db.reference("orders").get()

if orders:
    for order in orders:
        if order:
            cheapest = min(order["items"], key=lambda x: x["price"])
            print(f"Order {order['order_id']} - Cheapest Item:", cheapest)
else:
    print("No orders found in Firebase.")
```

```python
# Feature 4
orders = db.reference("orders").get()

if orders:
    customers = {order["customer"] for order in orders if order}
    for customer in customers:
        total_spent = sum(o["total_amount"] for o in orders if o and o["customer"] == customer)
        print(f"Customer {customer} - Total Spent:", total_spent)
else:
    print("No orders found in Firebase.")
```

```python
# Feature 5
orders = db.reference("orders").get()

if orders:
    product_counts = {}
    for order in orders:
        if order:
            for item in order["items"]:
                product_counts[item["product"]] = product_counts.get(item["product"], 0) + item["quantity"]

    most_ordered = max(product_counts, key=product_counts.get)
    least_ordered = min(product_counts, key=product_counts.get)

    print("Most Ordered Product:", most_ordered)
    print("Least Ordered Product:", least_ordered)
else:
    print("No orders found in Firebase.")
```

```python
# Feature 6
orders = db.reference("orders").get()

if orders:
    customer_counts = {}
    for order in orders:
        if order:
            customer_counts[order["customer"]] = customer_counts.get(order["customer"], 0) + 1

    multi_orders = [cust for cust, count in customer_counts.items() if count > 1]

    print("Customers with Multiple Orders:", multi_orders)
    if multi_orders:
        print("Customers with Multiple Orders:", multi_orders)
    else:
        print("No customers with multiple orders.")
else:
    print("No orders found in Firebase.")
```

```python
# Planned Outputs

orders = db.reference("orders").get()

if orders:
    orders = db.reference("orders").get()

    if orders:
        product_counts = {}
        for order in orders:
            if order:  # skip None entries
                for item in order["items"]:
                    product_counts[item["product"]] = product_counts.get(item["product"], 0) + item["quantity"]

        if product_counts:
            most_sold = max(product_counts, key=product_counts.get)
            least_sold = min(product_counts, key=product_counts.get)

            print("Most Sold Product:", most_sold)
            print("Least Sold Product:", least_sold)
        else:
            print("No products found in orders.")
    else:
        print("No orders found in Firebase.")

    avg_products = sum(
        sum(item["quantity"] for item in order["items"])
        for order in orders if order
    ) / len(orders)


    customer_counts = {}
    for order in orders:
        if order:
            customer_counts[order["customer"]] = customer_counts.get(order["customer"], 0) + 1

    avg_spend = sum(order["total_amount"] for order in orders if order) / len(customer_counts)


    customers_in_timeframe = len({
        order["customer"] for order in orders if order and "2025-08-01" <= order["order_date"] <= "2025-08-05"
    })

    print("Average Products per Order:", avg_products)
    print("Average Spend per Customer:", avg_spend)

    if customers_in_timeframe > 0:
        print("Customers served between 2025-08-01 and 2025-08-05:", customers_in_timeframe)
    else:
        print("No customers served in that timeframe.")
else:
    print("No orders found in Firebase.")
```
